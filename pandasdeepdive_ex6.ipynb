### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
# Set variables
s1 = sales1
s2 = sales2

# s2 cleaning and prep
col = s2.Title.str.extract(r'\(([^)]+)\)')
s2['Currency'] = col
s2 = s2.fillna(method='bfill')
s2 = s2[s2.Title.str.startswith('Sales report') != True]
s2 = s2[s2.Title.str.startswith('E-Book Reader') != True]
s2 = s2[s2.Title.str.contains('royalties') != True]

nc2 = s2.Title.str.upper()
s2['Book Title'] = nc2
s2 = s2.drop('Title', 1)

# s1 cleaning and prep
nc1 = s1['Book title'].str.upper()
s1['Book Title'] = nc1
s1 = s1.drop('Book title', 1)

# Merge
j = s2.merge(s1, on='Book Title')

# Consumer spending calc
j['Total_Sales'] = j['Units sold'] * j['Sales price']

# Final Summary
k = j.set_index(['Book Title', 'Currency'])
k = k.drop(['Units sold', 'List price', 'Royalty', 'Number sold', 'Sales price', 'Royalty paid'], 1)
k.unstack().fillna(0)
```
